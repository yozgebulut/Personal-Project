# Train - Test Split

```python
# ilgili kütüphanenin yüklenmesi
from sklearn.model_selection import train_test_split
```

```python
# sütunların eğitim ve test olarak ayrılması
X = df2.drop(columns = "YearsAtCompany")
y = df2["YearsAtCompany"]
```

```python
# verilerin %70 eğitim, %30 test için ayrılması
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# test ve eğitim için ayrılan verilerin kapsamına bakılması 
print("Train features shape : ", X_train.shape)
print("Train target shape   : ", y_train.shape)
print("Test features shape  : ", X_test.shape)
print("Test target shape    : ", y_test.shape)
```

# Fit Model

```python
# ilgili kütüphanelerin yüklenmesi
from sklearn.linear_model import LinearRegression
```

```python
# linear regression modelinin oluşturulması
model = LinearRegression()
```

```python
# linear regression modelinin verimizideki X ve Y' ye ait eğitim verilerine uydurulması
model.fit(X_train, y_train)
```

```python
# linear regression modeli tarafından yapılan tahminlerin bulunması
y_pred = model.predict(X_test)
```

# Handling And Plotting Residual

```python
# gerçek ve tahmin edilen değerler ile bu değerler arasındaki hata miktarının gösterilmesi
my_dict = {"Actual": y_test, "Pred": y_pred, "Residual":y_test - y_pred}
```

```python
# modelin performansının dataframe'e çevrilmesi
comparing = pd.DataFrame(my_dict)
comparing
```

```python
# test ve tahmin edilen değerler arasındaki farkın bulunması
residuals = y_test - y_pred
```

```python
# test ve residuals değerleri arasındaki farkın görselleştirilmesi
sns.scatterplot(x = y_test, y = residuals) 
plt.axhline(y = 0, color = "r", linestyle = "--")
plt.ylabel("Residuals")
plt.show()
```

# Error Metrics

```python
# ilgili kütüphanelerin yüklenmesi
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# error metriclerinin sonuçlarının gösterildiği bir fonksiyon tanımlanması
def train_val(model, X_train, y_train, X_test, y_test):
    
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    
    scores = {"TRAIN": {"R2 SCORE" : r2_score(y_train, y_train_pred),
    "MAE" : mean_absolute_error(y_train, y_train_pred),
    "MSE" : mean_squared_error(y_train, y_train_pred),                          
    "RMSE" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    "TEST": {"R2 SCORE" : r2_score(y_test, y_pred),
    "MAE" : mean_absolute_error(y_test, y_pred),
    "MSE" : mean_squared_error(y_test, y_pred),
    "RMSE" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
# error metriclerinin değerlerinin gösterilmesi
train_val(model, X_train, y_train, X_test, y_test)
```

- Modelin bağımsız değişkenler tarafından bağımlı değişkenin varyansının %72'sini açıkladı.

- Değerlerin 0 olması mükemmel tahminler yapıldığının göstergesidir.

- Değerlerin 0 ile 1 arasında olması başarılı bir modelin göstergesidir.

- Değerlerin 1 ile 2 arasında olması makul bir tahmin performansını göstergesidir. 

- Değerlerin 2'den büyük olması modelin tahminlerinin beklenenden daha fazla sapma gösterdiğinin göstergesidir.

```python
# test ve eğitim verilerine ait hata metriklerinin görselleştirilmesi
ax2 = train_val(model, X_train, y_train, X_test, y_test).plot(kind = "bar", title = ("Values of Error Metrics"),
                                                              color = ["skyblue", "blue"])
for p in ax2.patches:
    ax2.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha = "center", va = "center",
                 xytext = (0, 10), textcoords = "offset points")
plt.xticks(rotation = 45)

# png yi kaydetmek için
plt.savefig("error metrics",  bbox_inches = "tight")
```
